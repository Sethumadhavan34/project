**IMPORTING LIBRARIES**

```python
import csv 
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

LOADING DATA FROM SKLEARN

```python
df=sklearn.datasets.load_breast_cancer()
```

```python
df
```

**LOADING DATA TOA DATA FRAME**

```python
data_frame=pd.DataFrame(df.data,columns=df.feature_names)
```

PRINT FIRST 5 ROWS OF DATA FRAME


```python
data_frame.head()
```

**ADDING TARGET COLUMN TO DATA FRAME**

```python
data_frame["label"]=df.target
```

```python
#Print last 5 rows of a dataframe
data_frame.tail()
```

```python
#no.of rows and columns in dataset
data_frame.shape
```

```python
#ploting hist
data_frame.hist(figsize=(20,20))
```

```python
#Checking for the missing values
data_frame.isnull().sum()
```

```python
#Statistical measures about the data
data_frame.describe()
```

```python

#Checking the distribution of target variable
data_frame["label"].value_counts()
```

1-----> benign
0----->malignant


```python
data_frame.groupby("label").mean()
```

```python
x=data_frame.drop(columns="label",axis=1)
y=data_frame["label"]
```

```python
print(x)
```

```python
print(y)
```

**SPLITTING THE DATA INTO TRAINING DATA &TESTING DATA**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

MODEL TRAINING
LOGISTIC REGRESSION

```python
model=LogisticRegression()
```

```python
#Training the logistic regression model using training data
model.fit(x_train,y_train)
```

**MODEL EVALUATION**

**ACCURACY SCORE**

```python
#accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(y_train,x_train_prediction)
```

```python
print("accuracy on training data= ",training_data_accuracy)
```

```python
#accuracy on test data

x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(y_test,x_test_prediction)
```

```python
print("accuracy on test data= ",test_data_accuracy)
```
